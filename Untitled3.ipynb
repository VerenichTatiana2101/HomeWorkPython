```python
import pandas as pd
```

```python
import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI': lst})
data.head(20)
```

```python
unique_values = data['whoAmI'].unique()
one_hot = pd.DataFrame(0, index=data.index, columns=unique_values)
for i, value in enumerate(data['whoAmI']):
    one_hot.loc[i, value] = 1

one_hot.head(20)
```

```python
unique_values = []
for value in data['whoAmI']:
    if value not in unique_values:
        unique_values.append(value)

one_hot = []
for i in range(len(data)):
    row = []
    for j in range(len(unique_values)):
        if data['whoAmI'][i] == unique_values[j]:
            row.append(1)
        else:
            row.append(0)
    one_hot.append(row)

    print(*row)
```
